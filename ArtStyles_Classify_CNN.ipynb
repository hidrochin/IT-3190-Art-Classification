```python
import os

def print_total_pictures(base_folder):
    data_folder = os.path.join(base_folder, "Data")

    for split_type in ["train", "test", "val"]:
        split_folder_path = os.path.join(data_folder, split_type)
        print(f"\nTotal pictures in {split_type} folder:")

        for art_style in art_styles:
            art_style_folder_path = os.path.join(split_folder_path, art_style)
            total_pictures = sum(len(files) for _, _, files in os.walk(art_style_folder_path))
            print(f"{art_style}: {total_pictures} pictures")

# Example usage:
project_folder = "/content/drive/MyDrive/MachineLearning"
print_total_pictures(project_folder)
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

def plot_total_pictures(base_folder):
    data_folder = os.path.join(base_folder, "Data")

    # Update the art_styles list with your 5 art styles
    art_styles = ["rococo", "realism", "surrealism", "baroque", "neoclassicism"]
    bar_width = 1  # Increase the bar width for better visibility

    # Set up colors for train, test, and val
    colors = {'train': 'blue', 'val': 'green', 'test': 'red'}

    for i, art_style in enumerate(art_styles):
        total_art_style_pictures = 0  # Initialize total number of pictures for the art style
        max_heights = [0, 0, 0]  # Initialize a list to store heights for each column
        for j, split_type in enumerate(["train", "val", "test"]):
            split_folder_path = os.path.join(data_folder, split_type, art_style)
            total_pictures = sum(len(files) for _, _, files in os.walk(split_folder_path))

            # Plotting with different colors for each split type
            bar = plt.bar(i * (len(["train", "val", "test"]) + 1) + j * bar_width, total_pictures,
                          width=bar_width, label=f"{art_style.capitalize()}",
                          color=colors[split_type])

            # Add text label for total pictures above each group of bars
            plt.text(bar[0].get_x() + bar[0].get_width() / 2,
                     bar[0].get_height() + 5,
                     f'{total_pictures}', ha='center', va='bottom', fontsize=9, fontweight='bold', color='black')

            total_art_style_pictures += total_pictures  # Update total number of pictures for the art style
            max_heights[j] = max(max_heights[j], bar[0].get_height())

        # Add text label for total pictures below each group of bars
        plt.text(i * (len(["train", "val", "test"]) + 1) + (len(["train", "val", "test"]) * bar_width) / 2,
                 -20,
                 f'Total: {total_art_style_pictures}', ha='center', va='bottom', fontweight='bold', color='black', fontsize=10)

    plt.xlabel("Art Style")
    plt.ylabel("Total Pictures")
    plt.title("Total Pictures in Different Art Styles and Splits")
    plt.xticks(np.arange(len(art_styles)) * (len(["train", "val", "test"]) + 1) + (len(["train", "val", "test"]) * bar_width) / 2,
               [art_style.capitalize() for art_style in art_styles],
               rotation=45, ha='right')

    # Increase the figure size for better visibility
    plt.gcf().set_size_inches(12, 10)

    # Add legend outside the plot
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    # Add legend for split types
    plt.legend(["Train", "Validation", "Test"], loc='upper left', bbox_to_anchor=(1, 0.85))

    # plt.savefig('/content/drive/MyDrive/MachineLearning/Images/Data Visualization.png', bbox_inches='tight')

    plt.show()

# Example usage:
project_folder = "//content/drive/MyDrive/MachineLearning"
plot_total_pictures(project_folder)
```

## Image classification with Convolution Neural Network and Data Augumentation

Data augumentation

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define constants
image_width, image_height = 128, 128
batch_size = 32
num_classes = 5

# Define the paths to your data
train_dir = '/content/drive/MyDrive/MachineLearning/Data/train'
test_dir = '/content/drive/MyDrive/MachineLearning/Data/test'
validation_dir = '/content/drive/MyDrive/MachineLearning/Data/val'

# all images will be rescaled by 1/255.
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

# Create data generators for training, testing, and validation
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(image_width, image_height),
    batch_size=batch_size,
    class_mode='categorical',
    classes=['rococo', 'realism', 'surrealism', 'baroque', 'neoclassicism'],  # Specify the order of the classes
    shuffle=True
)

# Define your ImageDataGenerator with any necessary preprocessing options
datagen = ImageDataGenerator(rescale=1./255)

test_generator = datagen.flow_from_directory(
    test_dir,
    target_size=(image_width, image_height),
    batch_size=batch_size,
    class_mode='categorical',
    classes=['rococo', 'realism', 'surrealism', 'baroque', 'neoclassicism'],
    shuffle=False  # Do not shuffle test data
)

validation_generator = datagen.flow_from_directory(
    validation_dir,
    target_size=(image_width, image_height),
    batch_size=batch_size,
    class_mode='categorical',
    classes=['rococo', 'realism', 'surrealism', 'baroque', 'neoclassicism'],
    shuffle=True
)
```

### SVM

 Model for SMV using L1 Norm

```python
import tensorflow as tf
from tensorflow.keras.regularizers import l1

model_l1 = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 128x128 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    # The second convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The third convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The fourth convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),
    #Output layers for SVM
    tf.keras.layers.Dense(5, kernel_regularizer=l1(0.01), activation='softmax')
])

# Print the model summary
model_l1.summary()
```

 Model for SVM using L2 Norm

```python
import tensorflow as tf
from tensorflow.keras.regularizers import l2

model_l2 = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 128x128 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    # The second convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The third convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The fourth convolution
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),
    #Output layers for SVM
    tf.keras.layers.Dense(5, kernel_regularizer=l2(0.01), activation='softmax')
])

# Print the model summary
model_l2.summary()
```

```python
# Set the training parameters
model_l1.compile(loss = 'categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model_l2.compile(loss = 'categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

L1 Norm

```python
# Train the model
history = model_l1.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator,
    steps_per_epoch=len(train_generator),
    validation_steps=len(validation_generator),
    verbose=1
)

# Evaluate on the test set
test_results = model_l1.evaluate(test_generator, steps=len(test_generator))
print("Test loss:", test_results[0])
print("Test accuracy:", test_results[1])
```

Visualyze input through CNN

```python
import numpy as np
import random
import os
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model

# Assuming model_l1 is your original model
successive_outputs = [layer.output for layer in model_l1.layers[1:]]
visualization_model_l1 = Model(inputs=model_l1.input, outputs=successive_outputs)

# Define the paths to your data
train_dir = '/content/drive/MyDrive/MachineLearning/Data/train'
class_names = sorted(os.listdir(train_dir))

# Choose a random class
random_class = random.choice(class_names)
class_dir = os.path.join(train_dir, random_class)

# Get a list of image files in the chosen class
image_files = [os.path.join(class_dir, f) for f in os.listdir(class_dir) if f.endswith(".png")]

# Choose a random image from the class
img_path = random.choice(image_files)
img = load_img(img_path, target_size=(128, 128))  # This is a PIL image
x = img_to_array(img)  # Numpy array with shape (128, 128, 3)
x = x.reshape((1,) + x.shape)  # Numpy array with shape (1, 128, 128, 3)

# Scale by 1/255
x /= 255.0

# Run the image through the network, obtaining all intermediate representations for this image
successive_feature_maps = visualization_model_l1.predict(x)

# These are the names of the layers, so you can have them as part of your plot
layer_names = [layer.name for layer in model_l1.layers]

# Accumulate visualizations into a list
all_visualizations = []

# Display the representations
for layer_name, feature_map in zip(layer_names, successive_feature_maps):
    if len(feature_map.shape) == 4:
        # -------------------------------------------
        # Just do this for the conv / max-pool layers, not the fully-connected layers
        # -------------------------------------------
        n_features = feature_map.shape[-1]  # Number of features in the feature map
        size = feature_map.shape[1]  # Feature map shape (1, size, size, n_features)

        # Tile the images in this matrix
        display_grid = np.zeros((size, size * n_features))

        # -------------------------------------------------
        # Postprocess the feature to be visually palatable
        # -------------------------------------------------
        for i in range(n_features):
            x = feature_map[0, :, :, i]
            x -= x.mean()
            x /= x.std()
            x *= 64
            x += 128
            x = np.clip(x, 0, 255).astype('uint8')
            display_grid[:, i * size: (i + 1) * size] = x  # Tile each filter into a horizontal grid

        # Add the visualization to the list
        all_visualizations.append((layer_name, display_grid))

# Save all visualizations into one image
fig, axs = plt.subplots(len(all_visualizations), figsize=(20, 10))
for i, (layer_name, display_grid) in enumerate(all_visualizations):
    axs[i].imshow(display_grid, aspect='auto', cmap='viridis')
    axs[i].set_title(layer_name)
    axs[i].axis('off')

plt.savefig('/content/drive/MyDrive/MachineLearning/Images/CNN_Effect_All.png', bbox_inches='tight')
plt.show()  # This line is added to display the plots
```

### **KNN** and **RF**


```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(image_width, image_height, 3))

# Remove the classification head
x = GlobalAveragePooling2D()(base_model.output)

# Create a model with the modified base model
model = Model(inputs=base_model.input, outputs=x)

# Extract features from the training set
train_features = model.predict(train_generator)

# Flatten the features for KNN
train_features_flattened = train_features.reshape(train_features.shape[0], -1)
```

KNN

```python
# Create and train the KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(train_features_flattened, train_generator.classes)

# Extract features from the validation set
validation_features = model.predict(validation_generator)

# Flatten the features for KNN
validation_features_flattened = validation_features.reshape(validation_features.shape[0], -1)

# Predict using KNN on the validation set
knn_predictions = knn_classifier.predict(validation_features_flattened)

# Evaluate accuracy
knn_accuracy = accuracy_score(validation_generator.classes, knn_predictions)
print(f"KNN Accuracy: {knn_accuracy}")
```

RF

```python
# Create and train the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(train_features_flattened, train_generator.classes)

# Predict using Random Forest on the validation set
rf_predictions = rf_classifier.predict(validation_features_flattened)

# Evaluate accuracy
rf_accuracy = accuracy_score(validation_generator.classes, rf_predictions)
print(f"Random Forest Accuracy: {rf_accuracy}")
```
