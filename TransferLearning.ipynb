## Visualize the data

This part of the project is used for visualyzing data which are collected and settled in structure of the project.

```python
import os

def print_total_pictures(base_folder):
    data_folder = os.path.join(base_folder, "Data")

    for split_type in ["train", "test", "val"]:
        split_folder_path = os.path.join(data_folder, split_type)
        print(f"\nTotal pictures in {split_type} folder:")

        for art_style in art_styles:
            art_style_folder_path = os.path.join(split_folder_path, art_style)
            total_pictures = sum(len(files) for _, _, files in os.walk(art_style_folder_path))
            print(f"{art_style}: {total_pictures} pictures")

# Example usage:
project_folder = "/content/drive/MyDrive/MachineLearning"
print_total_pictures(project_folder)
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

def plot_total_pictures(base_folder):
    data_folder = os.path.join(base_folder, "Data")

    # Update the art_styles list with your 5 art styles
    art_styles = ["rococo", "realism", "surrealism", "baroque", "neoclassicism"]
    bar_width = 1  # Increase the bar width for better visibility

    # Set up colors for train, test, and val
    colors = {'train': 'blue', 'val': 'green', 'test': 'red'}

    for i, art_style in enumerate(art_styles):
        total_art_style_pictures = 0  # Initialize total number of pictures for the art style
        max_heights = [0, 0, 0]  # Initialize a list to store heights for each column
        for j, split_type in enumerate(["train", "val", "test"]):
            split_folder_path = os.path.join(data_folder, split_type, art_style)
            total_pictures = sum(len(files) for _, _, files in os.walk(split_folder_path))

            # Plotting with different colors for each split type
            bar = plt.bar(i * (len(["train", "val", "test"]) + 1) + j * bar_width, total_pictures,
                          width=bar_width, label=f"{art_style.capitalize()}",
                          color=colors[split_type])

            # Add text label for total pictures above each group of bars
            plt.text(bar[0].get_x() + bar[0].get_width() / 2,
                     bar[0].get_height() + 5,
                     f'{total_pictures}', ha='center', va='bottom', fontsize=9, fontweight='bold', color='black')

            total_art_style_pictures += total_pictures  # Update total number of pictures for the art style
            max_heights[j] = max(max_heights[j], bar[0].get_height())

        # Add text label for total pictures below each group of bars
        plt.text(i * (len(["train", "val", "test"]) + 1) + (len(["train", "val", "test"]) * bar_width) / 2,
                 -20,
                 f'Total: {total_art_style_pictures}', ha='center', va='bottom', fontweight='bold', color='black', fontsize=10)

    plt.xlabel("Art Style")
    plt.ylabel("Total Pictures")
    plt.title("Total Pictures in Different Art Styles and Splits")
    plt.xticks(np.arange(len(art_styles)) * (len(["train", "val", "test"]) + 1) + (len(["train", "val", "test"]) * bar_width) / 2,
               [art_style.capitalize() for art_style in art_styles],
               rotation=45, ha='right')

    # Increase the figure size for better visibility
    plt.gcf().set_size_inches(12, 10)

    # Add legend outside the plot
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

    # Add legend for split types
    plt.legend(["Train", "Validation", "Test"], loc='upper left', bbox_to_anchor=(1, 0.85))

    # plt.savefig('/content/drive/MyDrive/MachineLearning/Images/Data Visualization.png', bbox_inches='tight')

    plt.show()

# Example usage:
project_folder = "//content/drive/MyDrive/MachineLearning"
plot_total_pictures(project_folder)
```

```python
import os
import matplotlib.pyplot as plt
import random

def plot_sample_images(base_folder):
    data_folder = os.path.join(base_folder, "Data")

    # Update the art_styles list with your 5 art styles
    art_styles = ["rococo", "realism", "surrealism", "baroque", "neoclassicism"]

    fig, axs = plt.subplots(len(art_styles), 5, figsize=(15, 8), constrained_layout=True)

    for i, art_style in enumerate(art_styles):
        sample_folder_path = os.path.join(data_folder, "train", art_style)
        all_images = [os.path.join(sample_folder_path, f) for f in os.listdir(sample_folder_path)]

        # Randomly select 5 images
        sample_images = random.sample(all_images, min(5, len(all_images)))

        for j, (image_path, ax) in enumerate(zip(sample_images, axs[i])):
            img = plt.imread(image_path)
            ax.imshow(img)
            ax.axis("off")
            ax.set_title(f"{art_style.capitalize()} - {j + 1}", fontsize=10)

    # Save the figure before displaying it
    plt.savefig('/content/drive/MyDrive/MachineLearning/Images/Data_Sample.png', bbox_inches='tight')
    plt.show()

# Example usage:
project_folder = "/content/drive/MyDrive/MachineLearning"
plot_sample_images(project_folder)
```

## Transfer learning and Fintune Models with Restnet50

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
BATCH_SIZE = 32
IMG_SIZE = (128, 128)
initial_epoch = 10

train_dir = '/content/drive/MyDrive/MachineLearning/Data/train'
test_dir = '/content/drive/MyDrive/MachineLearning/Data/test'
validation_dir = '/content/drive/MyDrive/MachineLearning/Data/val'

train_dataset = tf.keras.utils.image_dataset_from_directory(train_dir,
                                                            shuffle=True,
                                                            batch_size=BATCH_SIZE,
                                                            image_size=IMG_SIZE)

test_dataset = tf.keras.utils.image_dataset_from_directory(test_dir,
                                                            shuffle=True,
                                                            batch_size=BATCH_SIZE,
                                                            image_size=IMG_SIZE)

validation_dataset = tf.keras.utils.image_dataset_from_directory(validation_dir,
                                                            shuffle=True,
                                                            batch_size=BATCH_SIZE,
                                                            image_size=IMG_SIZE)
```

### Transfer model

```python
AUTOTUNE = tf.data.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)
```

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip('horizontal'),
  tf.keras.layers.RandomRotation(0.2),
])
```

```python
preprocess_input = tf.keras.applications.resnet50.preprocess_input
```

```python
rescale = tf.keras.layers.Rescaling(1./127.5, offset=-1)
```

```python
# Create the base model from the pre-trained model Resnet50
IMG_SHAPE = IMG_SIZE + (3,)
base_model = tf.keras.applications.resnet50.ResNet50(input_shape=IMG_SHAPE,
                                                    include_top=False,
                                                    weights='imagenet')
```

```python
image_batch, label_batch = next(iter(train_dataset))
feature_batch = base_model(image_batch)
print(feature_batch.shape)
```

```python
base_model.trainable = False
```

```python
base_model.summary()
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
feature_batch_average = global_average_layer(feature_batch)
print(feature_batch_average.shape)
```

```python
prediction_layer = tf.keras.layers.Dense(5, activation='softmax')
prediction_batch = prediction_layer(feature_batch_average)
print(prediction_batch.shape)
```

```python
inputs = tf.keras.Input(shape=(128, 128, 3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = base_model(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
model = tf.keras.Model(inputs, outputs)

model.summary()
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
base_learning_rate = 1e-4

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name='accuracy')]
)
```

```python
loss0, accuracy0 = model.evaluate(validation_dataset)
```

```python
print("initial accuracy: {:.2f}".format(accuracy0))
```

```python
history = model.fit(train_dataset,
                    epochs=initial_epoch,
                    validation_data=validation_dataset)
```

```python
# Evaluate on the test set
test_results_trans = model.evaluate(test_dataset, steps=len(test_dataset))
print("Test loss:", test_results_trans[0])
print("Test accuracy:", test_results_trans[1])
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')
# plt.savefig('/content/drive/MyDrive/MachineLearning/Images/TransferLearningAcc.png', bbox_inches='tight')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,2.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.savefig('/content/drive/MyDrive/MachineLearning/Images/TransferLearning.png', bbox_inches='tight')

plt.show()
```

### Finetuning

Freeze 100 layers

```python
base_model.trainable = True

# Let's take a look to see how many layers are in the base model
print("Number of layers in the base model: ", len(base_model.layers))

# Fine-tune from this layer onwards
fine_tune_at = 100

# Freeze all the layers before the `fine_tune_at` layer
for layer in base_model.layers[:fine_tune_at]:
  layer.trainable = False
```

```python
base_learning_rate = 1e-4

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name='accuracy')]
)
```

```python
fine_tune_epochs = 10
total_epochs =  initial_epoch + fine_tune_epochs

history_fine = model.fit(train_dataset,
                         epochs=total_epochs,
                         initial_epoch=history.epoch[-1],
                         validation_data=validation_dataset)
```

```python
acc += history_fine.history['accuracy']
val_acc += history_fine.history['val_accuracy']

loss += history_fine.history['loss']
val_loss += history_fine.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.ylim([0.4, 1])
plt.plot([initial_epoch-1,initial_epoch-1],
          plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
# plt.savefig('/content/drive/MyDrive/MachineLearning/Images/FintuningAcc_100.png', bbox_inches='tight')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.ylim([0, 2.0])
plt.plot([initial_epoch-1,initial_epoch-1],
         plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.savefig('/content/drive/MyDrive/MachineLearning/Images/Finetuning_100.png', bbox_inches='tight')
plt.show()
```

```python
# Evaluate on the test set
test_results_1 = model.evaluate(test_dataset, steps=len(test_dataset))
print("Test loss:", test_results_1[0])
print("Test accuracy:", test_results_1[1])
```
